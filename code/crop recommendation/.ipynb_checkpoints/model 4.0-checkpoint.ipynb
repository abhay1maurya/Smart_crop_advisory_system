```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import xgboost as xgb
import warnings

# Sklearn Imports
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from scipy.stats import randint, uniform

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# 1. Load Data
# ==============================================================================
file_path = 'crop_data_final_data.csv'

try:
    df = pd.read_csv(file_path)
    print(f"‚úÖ Data Loaded. Shape: {df.shape}")
except FileNotFoundError:
    print(f"‚ùå CRITICAL ERROR: '{file_path}' not found. Please run the data repair script first.")
    raise

# 2. Safety Check (Prevent "Class with 1 sample" crashes)
# ==============================================================================
class_counts = df['label'].value_counts()
rogue_classes = class_counts[class_counts < 2].index

if len(rogue_classes) > 0:
    print(f"‚ö†Ô∏è WARNING: Found {len(rogue_classes)} classes with < 2 samples. Dropping them to prevent crash.")
    df = df[~df['label'].isin(rogue_classes)].copy()
else:
    print("‚úÖ Data Check Passed: All classes have sufficient samples.")

def add_smart_features(data):
    df = data.copy()
    
    # 1. Nutrient Ratios (Crucial for biological distinction)
    # Avoid division by zero by adding a tiny epsilon (1e-5)
    df['total_nutrients'] = df['N'] + df['P'] + df['K'] + 1e-5
    df['N_ratio'] = df['N'] / df['total_nutrients']
    df['P_ratio'] = df['P'] / df['total_nutrients']
    df['K_ratio'] = df['K'] / df['total_nutrients']
    
    # 2. Climate Interaction
    # "Aridity Index" proxy: Rainfall / Temperature
    df['aridity_index'] = df['rainfall'] / (df['temperature'] + 1e-5)
    
    # "Water Stress": High Temp + Low Humidity = Stress
    df['water_stress'] = df['temperature'] * (100 - df['humidity'])
    
    return df

print("üß† Engineering new features...")
df_engineered = add_smart_features(df)

```

```python
# 3. Separate Features and Target
# ==============================================================================
# Update X to use the new features
X = df_engineered.drop('label', axis=1) # Use ALL features including new ones
y = df_engineered['label']

print(f"‚úÖ New Feature Count: {X.shape[1]} (Original: 7)")
print("   Added: N_ratio, P_ratio, K_ratio, aridity_index, water_stress")

# 4. Label Encoding
# ==============================================================================
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print(f"‚úÖ Target Encoded. Unique Crops: {len(le.classes_)}")

# 5. Stratified Split (80% Train / 20% Test)
# ==============================================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# 6. Scaling (Mandatory for KNN)
# ==============================================================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("‚úÖ Data Scaled & Split.")
print(f"   Training Data: {X_train_scaled.shape}")
print(f"   Testing Data:  {X_test_scaled.shape}")
```

```python
def create_optimized_models():
    """Defines the search space for Random Forest, KNN, and XGBoost"""
    print("‚öôÔ∏è Configuring Model Architectures...")

    # A. Random Forest (Wide Search Space)
    rf_param_dist = {
        'n_estimators': randint(200, 500),
        'max_depth': [15, 20, 25, 30, None],
        'min_samples_split': randint(2, 11),
        'min_samples_leaf': randint(1, 4),
        'max_features': ['sqrt', 'log2']
    }
    
    # B. KNN (Distance & Neighbors)
    knn_param_dist = {
        'n_neighbors': randint(5, 15),
        'weights': ['uniform', 'distance'],
        'p': [1, 2]  # 1=Manhattan, 2=Euclidean
    }

    # C. XGBoost (Gradient Boosting)
    xgb_param_dist = {
        'n_estimators': randint(150, 400),
        'max_depth': randint(4, 10),
        'learning_rate': uniform(0.01, 0.2), # Random float between 0.01 and 0.21
        'subsample': uniform(0.6, 0.4),      # Random float between 0.6 and 1.0
        'colsample_bytree': uniform(0.6, 0.4)
    }

    # Wrap in RandomizedSearchCV
    # n_iter=15 means we try 15 random combinations for each model (Fast & Effective)
    models = {
        'rf': RandomizedSearchCV(
            RandomForestClassifier(random_state=42, n_jobs=-1),
            rf_param_dist, n_iter=15, cv=3, scoring='accuracy', n_jobs=-1, verbose=1, random_state=42
        ),
        'knn': RandomizedSearchCV(
            KNeighborsClassifier(n_jobs=-1),
            knn_param_dist, n_iter=10, cv=3, scoring='accuracy', n_jobs=-1, verbose=1, random_state=42
        ),
        'xgb': RandomizedSearchCV(
            xgb.XGBClassifier(random_state=42, n_jobs=-1, eval_metric='mlogloss'),
            xgb_param_dist, n_iter=15, cv=3, scoring='accuracy', n_jobs=-1, verbose=1, random_state=42
        )
    }
    
    return models

model_searches = create_optimized_models()
print("‚úÖ Architectures Configured.")
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

def train_stacking_model(model_searches, X_train, y_train):
    print("üöÄ Starting Stacking Training (This is heavier than Voting)...")
    
    best_estimators = []
    
    # Phase 1: Tuning (Same as before)
    for name, search in model_searches.items():
        print(f"\nüîé Tuning {name.upper()}...")
        search.fit(X_train, y_train)
        print(f"   Best Score: {search.best_score_:.4f}")
        best_estimators.append((name, search.best_estimator_))
        
    print("\nüèóÔ∏è Building Stacking Architecture...")
    
    # Phase 2: Stacking
    # - Estimators: Your optimized RF, KNN, XGB
    # - Final Estimator: A Logistic Regression that learns how to weigh them
    stacking_clf = StackingClassifier(
        estimators=best_estimators,
        final_estimator=LogisticRegression(),
        cv=3,       # Internal Cross Validation to train the 'Manager'
        n_jobs=1    # CRITICAL: Keep as 1 to avoid Windows Crash
    )
    
    # Fit
    print("   Training Meta-Model...")
    stacking_clf.fit(X_train, y_train)
    
    return stacking_clf

# Execute
final_model = train_stacking_model(model_searches, X_train_scaled, y_train)
print("\n‚úÖ Stacking Model Trained.")
```

```python
# 1. Prediction
print("üìä Evaluating on Test Set...")
y_pred = final_model.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred)

print(f"\nüèÜ Final Test Accuracy: {acc*100:.2f}%")

# 2. Confusion Matrix Plot
print("\nüé® Plotting Confusion Matrix...")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=False, cmap='Reds', fmt='d')
plt.title(f'Confusion Matrix (Acc: {acc:.2f})')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# 3. Detailed Report
print("\nüìù detailed Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
# 1. Save Artifacts
joblib.dump(final_model, 'crop_model_optimized.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(le, 'label_encoder.pkl')

print("üíæ Success! Saved 'crop_model_optimized.pkl', 'scaler.pkl', and 'label_encoder.pkl'.")

# 2. Verification Function (Run this to test if the saved model works)
def test_prediction(N, P, K, temp, hum, ph, rain):
    # Load
    model = joblib.load('crop_model_optimized.pkl')
    sc = joblib.load('scaler.pkl')
    enc = joblib.load('label_encoder.pkl')
    
    # Prepare
    input_data = np.array([[N, P, K, temp, hum, ph, rain]])
    input_scaled = sc.transform(input_data)
    
    # Predict
    probs = model.predict_proba(input_scaled)
    best_idx = np.argmax(probs)
    label = enc.inverse_transform([best_idx])[0]
    conf = probs[0][best_idx] * 100
    
    return label, conf

# Test with "Rice-like" conditions
pred, confidence = test_prediction(80, 40, 40, 25, 80, 7, 200)
print(f"\nüß™ Test Prediction (Rice Conditions):")
print(f"   Predicted: {pred}")
print(f"   Confidence: {confidence:.2f}%")
```
